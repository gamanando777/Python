# 대표값

- 평균
- 중앙값
- 최빈값

```python
import numpy as np
import statistics as st
import pandas as pd
```

```python
print(dir(st))
```

```python
# 표본 생성 
# np.random.randint?
# np.random.randint(start, end, size=갯수)
data = np.random.randint(1, 100, 100)
data
```

```python
print(f' 평균 = {st.mean(data)}')
print(f' 중앙값 = {st.median(data)}')
print(f' 최빈값 = {st.mode(data)}')
# 리스트명.count(값)
print(f' 최빈값 갯수 = {list(data).count(st.mode(data))}')
```

## 분산
- variance()

## 표준편차
- stdev()


```python
data = np.random.randint(10, 41, size=50)
data
```

```python
print(f' 평균 = {st.mean(data)}')
print(f' 분산 = {st.variance(data)}')
print(f' 표준편차 = {st.stdev(data):.2f}')
print(f' 표준편차 = {round(st.stdev(data),2)}')
```
