```python
# 2. 2000년대 이후데이터를 구하시요
# 년도   남아   여아
# ==============
# 2000     xx     xx
print('년도','남아','여아',sep='\t')
print('='*30)
for y,m,f in birthArr[ birthArr[:,0] >=2000 ]:
    print(y,m,f)
```

```python
# 3. 남아출생이 가장많은 년도와 남아수를 구하시요.
mx = birthArr[:,1].max()
for y,m,f in birthArr[ birthArr[:,1] == mx ]:
    print('년도:',y, '남아수:',m)
# birthArr[birthArr[:,1] == 2044160]
```

```python
# 4. 남아출생이 가장많은 top5 를 구하시요(년도와 남아수)
a= np.argsort(birthArr[:,1], axis=0)[:-6:-1]
birthArr[a,:2]
```

```python
idx = np.argsort( birthArr[:,1] )[:-6:-1]
birthArr[ idx ][:, [0,1] ]
```

```python
5. 남아율(%) 을 구하고 각년도별
남아수가 차지하는비율을 데이터로
보여주시요.
남아    여아    남아율
===================
..
```

```python
m = birthArr[:,1]
w = birthArr[:,2]
ratio = (m/(m+w))*100
print('='*20)
print('남아','여아','남아율',sep='\t')
print('='*20)
for m,w,r in zip(m,w,ratio.round(2)):
    print(m, w, r,sep='\t')
```

```python
6. 2000년도 이후데이터 에 대해
출생량 을 구하고
남아수가 1800000 이상이면 '많음'아니면 '적음'
이라고 보여 주시요
	남아수 출생량
2000   1233   '많음'
2001   1234   '적음'
```

```python
birth20 = birthArr[birthArr[:,0] >= 2000]
result = np.where( birth20[:,1] > 1800000, '많음', '적음')

print('='*20)
print('연도','남아수','출생량',sep='\t')
print('='*20)
for a, b, c in zip(birth20[:,0], birth20[:,1], result):
    print(a, b, c, sep='\t')
```

```python
# 7. 1800년대,1900년대,2000년대 
# 각 남아,여야수 평균을 구하시요
```

```python
b1800=birthArr[birthArr[:,0]<1900]
b1900=birthArr[ (birthArr[:,0]>=1900) & (birthArr[:,0]<2000)]
b2000=birthArr [birthArr[:,0]>=2000]
y, m ,f = b1800.mean(axis=0)
print( '1800년대','남아평균:',m, '여야평균:',f)
y, m ,f = b1900.mean(axis=0)
print( '1900년대','남아평균:',m, '여야평균:',f)
y, m ,f = b2000.mean(axis=0).round(2)
print( '2000년대','남아평균:',m, '여야평균:',f)
```
