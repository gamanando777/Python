조선일보 RSS
- rss.joins.com

```python
jurl = 'https://rss.joins.com/joins_homenews_list.xml'
```

```python
response = REQ.urlopen(jurl)
rss = BeautifulSoup( response,'html.parser')
rss
```

```python
rss.select('item')
```

```python
for item in rss.select('item'):
    print('기사제목:',item.title.string)
    print('기사내용:',item.description.string)
    print('===================================')
```
