```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.plot( x, y1 ) #디폴트 창 출력
plt.figure(figsize=(3,3))  #창
plt.plot( x, y2 ) 
plt.plot( x, y3 ) 
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.subplot(2,2,1) # 2행 2열의 1번째
plt.plot( x, y1 )
plt.subplot(2,2,2) # 2행 2열의 2번째
plt.plot( x, y2 ) 
plt.subplot(2,2,3) #2행 2열의 3번째
plt.plot( x, y3 ) 
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.subplot(1,3,1) # 1행 3열의 1번째
plt.plot( x, y1 )
plt.subplot(1,3,2)
plt.plot( x, y2 ) 
plt.subplot(1,3,3)
plt.plot( x, y3 ) 
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
fig,axes = plt.subplots(2,2,figsize=(7,7)) #행,열, 사이즈
print(axes)
axes[0,0].plot(x,y1, 'ro-')
axes[0,1].plot(x,y2, 'bs--')
axes[1,0].plot(x,y3, 'g>-.')
```

```python
fig = plt.figure(figsize=(7,7))
ax1 = fig.add_subplot(2,2,1) # fig.add_subplot(221)
ax2 = fig.add_subplot(2,2,2) # fig.add_subplot(222)
ax3 = fig.add_subplot(2,2,3) # fig.add_subplot(223)

ax1.set_title('라인1')
ax1.set_xlabel('x축')
ax1.set_ylabel('y축')

ax2.set_title('라인2')
ax3.set_title('라인3')


ax1.plot( x,y1)
ax2.plot( x,y2)
ax3.plot( x,y3)
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]

fig,ax1 = plt.subplots() # 디폴트는 1행 1열. 지정하지 않는 것과 같다

ax1.set_xlabel('x축')
ax1.set_ylabel('y축',color='b')

ax1.plot(x,y1)

ax2 = ax1.twinx() #y축을 오른쪽에
ax2.set_ylabel('y2축',color='r')
ax2.plot(x,y2,'r--')
plt.show()
```
