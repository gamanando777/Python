```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import urllib.request as REQ

import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

https://www.menupan.com/

```python
murl = 'https://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=cn201'
```

```python
response = REQ.urlopen(murl)
soup = BeautifulSoup( response, 'html.parser')
soup
```

```python
sel = 'body > div > div.container > div.bestRest > div.ranking > div.rankingList > ul'
ul = soup.select_one(sel)
ul.select('li')
```

```python
for li in ul.select('li'):
    print('순위: ', li.select_one('.numTop,.rankNum').text)
    print('상호명: ', li.select_one('.restName > a').text)
    print('업종: ', li.select_one('.listType').text)
    print('지역: ', li.select_one('.listArea').text)    
    print('--------------------------------------')
```

```python
menuList=[]
for li in ul.select('li'):
    menuList.append({'순위':li.select_one('.numTop,.rankNum').text,
                    '상호명':li.select_one('.restName > a').text,
                    '업종':li.select_one('.listType').text,
                    '지역':li.select_one('.listArea').text})
    print('순위: ', li.select_one('.numTop,.rankNum').text)
    print('상호명: ', li.select_one('.restName > a').text)
    print('업종: ', li.select_one('.listType').text)
    print('지역: ', li.select_one('.listArea').text)    
    print('--------------------------------------')
```

```python
menuList
```

```python
menuDF = pd.DataFrame(menuList)
menuDF
```

```python
href=[]
for li in ul.select( 'li' ):
    href.append(  li.select_one( '.restName > a')['href'] )
```

```python
href
```

```python
tel = []
addr = []
for h in href:
    response = REQ.urlopen( 'https://www.menupan.com'+h )
    html = BeautifulSoup( response, 'html.parser')
    print( '전화번호:', html.select_one( '.tel1').string )
    print( '주소:', html.select_one( '.add1 > a').string )
    tel.append( html.select_one( '.tel1').string )
    addr.append( html.select_one( '.add1 > a').string )
```

```python
tel
```

```python
addr
```

```python
menuDF['전화번호'] = tel
menuDF['주소'] = addr
```
