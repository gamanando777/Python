```python
df2014=pd.read_csv('data/2014년 졸음운전 교통사고.csv',encoding='euc-kr')
df2015=pd.read_csv('data/2015년 졸음운전 교통사고.csv',encoding='euc-kr')
df2016=pd.read_csv('data/2016년 졸음운전 교통사고.csv',encoding='euc-kr')
```

```python
df2014 = pd.DataFrame(df2014)
df2015 = pd.DataFrame(df2015)
df2016 = pd.DataFrame(df2016)
```

```python
1. 3개의 데이터를 병합하시요.
====이후문제는 병합한 데이터프레임을 이용 ===========
```

```python
df=pd.concat([df2014,df2015,df2016])
df
```

```python
2. 년도, 월별로 인덱스를 설정(멀티인덱스) 하고
데이터를 보여 주시요
```

```python
df['년도'] = df['구분'].str[:5]
df['월'] = df['구분'].str[5:]
# df['구분'].str.extract( '([0-9]+)년([0-9]+)월')

```

```python
df.set_index(['년도','월'], inplace=True)
df.drop(columns='구분', inplace=True)
```

```python
# 3. 년도 및 월별 평균 사망자를 보여주시요
df.pivot_table(index = ['년도','월'])[['사망(명)']]
```

```python
4.  2016년 사고대비 사망율을 구하시요
```

```python
(df.loc['2016년']['사망(명)'].sum()/df.loc['2016년']['사고(건)'].sum()*100).round(2)
```

```python
5. 2014년도 월별 사망, 부상 데이터를 바차트로
보여주시요
```

```python
df.loc['2014년'].plot(kind='bar',y=['사망(명)','부상(명)'] )
plt.show()
```

```python
# 6. 2015년 대비  사망이 가장 많이 증가한
# 2016년 도 월을 구하시요

d16 = df.loc['2016년','사망(명)']
d15 = df.loc['2015년','사망(명)']
d16[(d16-d15) == (d16-d15).max()]
```
